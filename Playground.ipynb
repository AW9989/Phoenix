#### Basic CC Simulation 
Chen2020 is Parameter Set for a LGM50 cell (comprised of a NMC 811 positive electrode and bi-component Graphite-SiOx negative electrode)


```python
import pybamm 
model = pybamm.lithium_ion.DFN()
parameter_values = pybamm.ParameterValues("Chen2020") #Loading predefined Parameter Sets
#from Parameter_Sets.Parameters_Alex import get_parameter_values
#model.print_parameter_info() #Printing all the parameters the model needs



#Run the simulation 
sim = pybamm.Simulation(model, parameter_values=parameter_values)
sim.solve([0, 3600])
#sim.plot()
```

#### Complex Experimental Routines with simple model

```python
experiment = pybamm.Experiment(
    [
        (
            "Discharge at C/10 for 10 hours or until 3.3 V",
            "Rest for 1 hour",
            "Charge at 1 A until 4.1 V",
            "Hold at 4.1 V until 50 mA",
            "Rest for 1 hour",
        )
    ]
    * 1
    + [
        "Discharge at 1C until 3.3 V",
    ]
)
sim = pybamm.Simulation(model, parameter_values=parameter_values, experiment=experiment)
sim.solve()
#sim.plot()
```
